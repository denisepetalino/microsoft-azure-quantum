### Entanglement
States of different qubits depend on each other in such a way that the individual qubits cannot be described independently.

**Quantum teleportation** = where you transfer the state of a qubit from one location to another without physically moving the qubit itself. 
- use operations CNOT, H, Pauli X and Z to create entangled Bell states.

## Understanding quantum entanglement
Imagine A and B, two qubits completely independent from each other. Both have nothing to do with each other's state and are not entangled as they do not share any information with each other.

If you entangle the qubits, the state of A now depends on B and vice versa (now correlated).



When two particles become entangled, we can no longer descreibe them independently. We must now describe the particles together as a whole system.

## Description of quantum entanglement

But what is the individual state of qubit A and of B? Here, you cannot describe the state of qubit A without considering the state of qubit B as the individual qubit systems A and B are entangled. -> correlated and cannot be described independently.



In the language of linear algebra, a multi-qubit system is entangled when the system can't be constructed as a tensor product of the individual qubit states. A good exercise is to try to decompose the two-qubit system state into a tensor product of the state of qubit A and the state of qubit B. You find that it's impossible to do so.



In +, if you measure qubit A and get 0, you immediately know qubit B must be 0 (and similarly for 1) because the mismatched outcomes are impossible.

In the Bell state:



- amplitude of 00 is 1/root 2
- amplitude of 11 is 1/root 2
- amplitude of 01 is 0
- amplitude of 10 is 0

Measurement amplitudes come from the Born rule: P(outcome) = (amplitude)^2

So:

- P(00) = (1/root 2)^2 = 1/2
- P(11) = (1/root 2)^2 = 1/2
- P(01) = 0^2 = 0
- P(10) = 0^2 = 0

This is why 01 and 10 never occur for this state -> amplitudes are exactly zero.

### The Controlled-NOT (CNOT) operation
operation that helps create entanglement.

CNOT takes two qubits as input:  control qubit and target qubit.

If control qubit is in 1> state, then the CNOT operation flips the state of the target qubit. Otherwise CNOT does nothing.



In Q# CNOT acts on array of two qubits. [0] = control [1] = target

### How to use entanglement to send information
Quantum teleportation uses entanglement to transfer the state of a qubit from one location to another. State of one qubit is transferred to another qubit, but the qubit itself does not physically move.

Teleportation protocol uses entanglement and classical communication (as teleportation protocol requires the sender to communictae the results of their measurements to the receiver).
