### What is resouce estimation
Estimating the resources required for a quantum algorithm would need to run with the assumption it is being run on a fault-tolerant, large-scale quantum computer:
- no of qubits
- no of quantum gates
- processing time

Resource requirements can vary signigicantly depending on:
- qubit technology
- error correction scheme

AES requires significantly higher qubits, runtime than RSA and Elliptic Curve. Resources are higher for superconducting qubits than for topological qubits.

### Choose the error budget
- total error budget sets the allowed overall failure probability for the algorithm
- default is 0.001 so algorithm is allowed to fail once per 1000 runs
- appropriate error budget depends on the application:
- shor's algorithm: larger error can be acceptable as you can verify the factoring result efficiently
- for problems where solutions are not easily verifiable, may need a smaller error budget
